**Create General Expected Score Function**

Offense <br><br>

Variables --> 


* Drives Per Game   
* TDs % of all Drives 
* FG % of all Drives
* Scoreless % of all Drives

```python
def exp_FB_Score():

  a_drives = [0,3,7]  

  drive_Input = 12

  drives = round(drive_Input)

  scoreless_Percent = .6

  fg_Percent = .28

  td_Percent = .12


  l = random.choices(a_drives, weights=
                     [scoreless_Percent, fg_Percent, td_Percent], k=drives)


  a0 = 0
  a3 = 0
  a7 = 0

  for x in l:
    if x == 0:
      a0 += 1
  for x in l:
    if x == 3:
      a3 += 1
  for x in l:
    if x == 7:
      a7 += 1

  exp_Score = (a0 * 0) + (a3 * 3) + (a7 * 7)


  return exp_Score

exp_FB_Score()
```

```python
def exp_FB_Score():

  off_drives = [0,3,7]  

  off_Drive_Input = 12

  off_Drives = round(off_Drive_Input)

  off_Scoreless_Percent = .6

  off_Fg_Percent = .28

  off_Td_Percent = .12


  off_Scores = random.choices(off_drives, weights=
                     [off_Scoreless_Percent, off_Fg_Percent, off_Td_Percent], 
                     k=off_Drives)


  off_0 = 0
  off_3 = 0
  off_7 = 0

  for x in off_Scores:
    if x == 0:
      off_0 += 1
  for x in off_Scores:
    if x == 3:
      off_3 += 1
  for x in off_Scores:
    if x == 7:
      off_7 += 1

  off_Exp_Score = (off_0 * 0) + (off_3 * 3) + (off_7 * 7)


  def_drives = [0,3,7]  

  def_Drive_Input = 10

  def_Drives = round(off_Drive_Input)

  def_Scoreless_Percent = .65

  def_Fg_Percent = .25

  def_Td_Percent = .1


  def_Scores = random.choices(def_drives, weights=
                     [def_Scoreless_Percent, def_Fg_Percent, def_Td_Percent],
                      k=def_Drives)


  def_0 = 0
  def_3 = 0
  def_7 = 0

  for x in def_Scores:
    if x == 0:
      def_0 += 1
  for x in def_Scores:
    if x == 3:
      def_3 += 1
  for x in def_Scores:
    if x == 7:
      def_7 += 1

  def_Exp_Score = (def_0 * 0) + (def_3 * 3) + (def_7 * 7)

  exp_Score_Spread = (abs((off_Exp_Score - def_Exp_Score)/2))

  middle_Of_Both = (exp_Score_Spread + 
                    min(off_Exp_Score, def_Exp_Score))


  #return off_Exp_Score, def_Exp_Score, middle_Of_Both, exp_Score_Spread

  return middle_Of_Both

exp_FB_Score()






```

**Create Mass Sim of Expected Score**
<br><br><br>
Variable

```python
def exp_FB_Score_Mass_list(x):

  collection = []
  total = 0

  # Average with 12 Drives, .6, .28, & .12 --> 20.15 pts


  for i in range(x):
    collection.append(exp_FB_Score())

  for l in collection:
    total += l

  return [collection]


exp_FB_Score_Mass_list(15)
```

**Create Mass Sim of AVG Expected Score**
<br><br><br>
Variable

```python
def exp_FB_Score_Mass_Avg(x):

  collection = []
  total = 0

  # Average with 12 Drives, .6, .28, & .12 --> 20.15 pts

  for i in range(x):
    collection.append(exp_FB_Score())

  for l in collection:
    total += l

  scores = len(collection)

  average = (total/scores)

  return average

exp_FB_Score_Mass_Avg(10000)
```

**Create Mass Sim of Expected Score O/U **
<br><br><br>
Variable

```python
def exp_FB_Score_Mass_Avg_OU(x):

  collection = []
  total = 0

  # Average with 12 Drives, .6, .28, & .12 --> 20.15 pts

  bar = 19.999

  over = 0
  under = 0

  for i in range(x):
    collection.append(exp_FB_Score())

  for l in collection:
    total += l


  for x in collection:
    if x > bar:
      over += 1
    if x < bar:
      under += 1

  return over, under

  #print(average)

exp_FB_Score_Mass_Avg_OU(100000)
```
